# Problem Statement

Many businesses struggle to understand their diverse customer base and optimize their marketing efforts. This lack of insight often leads to inefficient use of resources, customer dissatisfaction, and missed opportunities for increased revenue. To address these issues, the company seeks a systematic approach to identify and categorize customers based on their purchasing behaviors to enhance customer engagement, retention, and overall profitability.

# Objectives

1. **Identify Key Customer Segments**: Use RFM analysis to segment the customer base into meaningful groups, such as loyal customers, big spenders, and at-risk customers.

2. **Improve Customer Retention**: Develop targeted strategies to retain high-value customers and re-engage those at risk of churning.

3. **Increase Customer Lifetime Value (CLV)**: Focus on strategies that increase the frequency and value of customer transactions.

4. **Optimize Marketing Strategies**: Tailor marketing efforts based on customer segments to maximize the effectiveness of campaigns and promotions.

5. **Enhance Customer Experience**: Use insights from RFM analysis to provide a more personalized and engaging customer experience.

# Procedures

1. **Data Collection:** Gather customer transaction data, including purchase dates, frequency of purchases, and monetary value of each transaction.
2. **RFM Scoring:**
   - **Recency (R):** Score customers based on how recently they made a purchase.
   - **Frequency (F):** Score based on how often they purchase.
   - **Monetary (M):** Score based on the total amount they have spent.
3. **Segmentation:** Combine R, F, and M scores to categorize customers into segments such as:
   - **Loyal Customers:** High R, F, and M scores.
   - **Big Spenders:** High M, moderate R and F scores.
   - **Going Cold:** High R, low F, and M scores.
   - **Recent Regulars:** Low R, moderate F and M scores.
4. **Strategy Development:** Create targeted strategies for each segment, such as loyalty programs for loyal customers, exclusive offers for big spenders, and reactivation campaigns for those going cold.
5. **Implementation and Monitoring:** Implement the tailored strategies and monitor their effectiveness through customer feedback and key performance indicators (KPIs) such as retention rates, average order value, and customer satisfaction scores.

# RFM Analysis

### Creating the RFM DataFrame

```python
data1 = data.copy()
```

```python
import datetime as dt
```

```python
# Define the reference date as the day after the last transaction
reference_date = data1['trans_date'].max() + dt.timedelta(days=1)
```

```python
# Calculate RFM metrics for each customer
rfm = data1.groupby('cust_id').agg({
    'trans_date': lambda x: (reference_date - x.max()).days,  # Recency
    'cust_id': 'count',                                        # Frequency
    'amount': 'sum'                                            # Monetary
}).rename(columns={
    'trans_date': 'Recency',
    'cust_id': 'Frequency',
    'amount': 'Total Amount'
}).reset_index()
```

```python
rfm.head()
```

```python
rfm['Monetary'] = rfm['Total Amount'] / rfm['Frequency']
rfm.head()
```

```python
rfm = rfm.drop(columns=['Total Amount'], axis=1)
rfm.head()
```

```python
rfm.shape
```

```python
rfm1=rfm.copy(deep=True)
```

### Clustering with K-Means

```python
clusters = int(input("Enter the maximum number of clusters: "))
```

```python
# Using the elbow method to find the optimal number of clusters
sse = []
for k in range(1, clusters+1):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_scaled)
    sse.append(kmeans.inertia_)
```

```python
# Calculate differences between successive SSE values
sse_diff = [j - i for i, j in zip(sse[:-1], sse[1:])]
```

```python
tolerence = float(input("Enter the tolerence limit: "))
```

```python
# Determine the optimal number of clusters
optimal_k = next((i+2 for i, diff in enumerate(sse_diff) if diff < tolerence), len(sse))
optimal_k
```

```python
# Plotting the elbow curve with SSE values
plt.figure(figsize=(10, 6))
plt.plot(range(1, clusters+1), sse, marker='o')

# Annotating each point with the SSE value
for i, txt in enumerate(sse):
    plt.annotate(f'{txt:.2f}', (i+1, sse[i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.xlabel('Number of clusters')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.axvline(x=optimal_k, color='r', linestyle='--', label=f'Optimal k = {optimal_k}')
plt.legend()
plt.show()
```

```python
# Choose the number of clusters (k) based on the elbow curve
kmeans = KMeans(n_clusters = optimal_k, random_state=42)
rfm['Cluster'] = kmeans.fit_predict(data_scaled)

# Inspect the clusters
print(rfm.groupby('Cluster').mean())
```

```python
# Step 1: Define custom bins with sufficient range
recency_bins = [0, 365, 1070, rfm1['Recency'].max() + 1]  # Include all possible values
frequency_bins = [0, 2, 4, rfm1['Frequency'].max() + 1]    # Include all possible values
monetary_bins = [0, 50, 100, rfm1['Monetary'].max() + 1]    # Include all possible values
```

```python
# Step 2: Assign scores using pd.cut
rfm1['R_Score'] = pd.cut(rfm1['Recency'], bins=recency_bins, labels=[3, 2, 1], right=False)
rfm1['F_Score'] = pd.cut(rfm1['Frequency'], bins=frequency_bins, labels=[1, 2, 3], right=False)
rfm1['M_Score'] = pd.cut(rfm1['Monetary'], bins=monetary_bins, labels=[1, 2, 3], right=False)

# Step 3: Check for NaNs
print("NaNs in R_Score:", rfm1['R_Score'].isna().sum())
print("NaNs in F_Score:", rfm1['F_Score'].isna().sum())
print("NaNs in M_Score:", rfm1['M_Score'].isna().sum())
```

```python
# Recalculate total RFM score
rfm1['RFM_Score'] = rfm1['R_Score'].astype(int) * 100 + rfm1['F_Score'].astype(int) * 10 + rfm1['M_Score'].astype(int)

# Check the distribution of RFM scores
rfm_score_counts = rfm1['RFM_Score'].value_counts().sort_index()
print("RFM Score Counts:\n", rfm_score_counts)
```

```python
# Convert to DataFrame for easier handling
rfm_df = pd.DataFrame(rfm_score_counts).reset_index()
rfm_df.columns = ['RFM_Score', 'Count']

# Step 2: Sort the DataFrame by 'Count' in descending order
rfm_df = rfm_df.sort_values(by='Count', ascending=False)
```
